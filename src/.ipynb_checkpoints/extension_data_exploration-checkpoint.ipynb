## Data Filtering: McMinnville Business Registration

In this code snippet,  `mcminnville_data` is a DataFrame containing business registration data for McMinnville, several steps are taken to filter and display relevant information. First, the 'Registry Date' column is converted to the datetime format using Pandas' `pd.to_datetime` function, ensuring proper treatment of date information. Next, a start date (e.g., '1980-01-01') and the current date are defined. The DataFrame is then filtered to include only businesses registered between this start date and the current date. The resulting `filtered_data` DataFrame showcases businesses registered within the specified time frame. Notably, the `errors='coerce'` parameter is used during the datetime conversion to handle any parsing errors gracefully, setting them as NaT (Not a Time) and allowing the code to proceed uninterrupted. This process aids in extracting and examining business registration data for McMinnville within a specified temporal range.


```python

# Assuming df is your DataFrame

# Convert 'Registry Date' to datetime format
mcminnville_data['Registry Date'] = pd.to_datetime(mcminnville_data['Registry Date'], errors='coerce')

# Filter businesses registered from 1980 to the current date
start_date = pd.to_datetime('1980-01-01')  # Adjust the start date as needed
current_date = pd.to_datetime('today')

filtered_data = mcminnville_data[(mcminnville_data['Registry Date'] >= start_date) & (mcminnville_data['Registry Date'] <= current_date)]

# Display the filtered data
print(filtered_data)
```

## Year-wise Business Registration Analysis

This code segment focuses on conducting a year-wise analysis of business registrations in McMinnville. Initially, it checks whether the 'Registry Date' column in the previously filtered data (`filtered_data`) is in datetime format. If not, it converts it to datetime using the `pd.to_datetime` function with `errors='coerce'` to handle any potential parsing errors. Subsequently, a new column named 'Year' is created, extracting the year information from the 'Registry Date' column using Pandas' `dt.year` attribute. This 'Year' column facilitates the grouping of business registrations by year. The code then performs a year-wise analysis, exemplified by the count of businesses registered each year. The results are stored in the `yearly_analysis` DataFrame and displayed to provide insights into the temporal distribution of business registrations in McMinnville.


```python

# Convert 'Registry Date' to datetime format (if not already done)
filtered_data['Registry Date'] = pd.to_datetime(filtered_data['Registry Date'], errors='coerce')

# Extract the year from the 'Registry Date' column
filtered_data['Year'] = filtered_data['Registry Date'].dt.year

# Perform year-wise analysis (example: count of businesses registered each year)
yearly_analysis = filtered_data['Year'].value_counts().sort_index()

# Display the year-wise analysis
print("Year-wise Analysis:")
print(yearly_analysis)
```

```python

# Assuming df is your DataFrame with the filtered data
# Replace 'Entity Type' with the actual column name if it's different in your DataFrame
entity_type_column = 'Entity Type'
registry_date_column = 'Registry Date'

# Convert 'Registry Date' to datetime format if not already done
mcminnville_data[registry_date_column] = pd.to_datetime(mcminnville_data[registry_date_column], errors='coerce')

# Group by 'Entity Type' and 'Registry Date', count the occurrences
grouped_data = mcminnville_data.groupby([entity_type_column, mcminnville_data[registry_date_column].dt.year])[entity_type_column].count().unstack().T

# Filter entity types based on the trend (you can adjust the threshold as needed)
filtered_entity_types = []
for entity_type in grouped_data.columns:
    # Calculate the slope of the linear regression line
    slope = grouped_data[entity_type].diff().mean()
    
    # Filter based on slope threshold
    if slope <= 0.5:  # You can adjust this threshold as needed
        filtered_entity_types.append(entity_type)

# Plotting only selected entity types
for entity_type in filtered_entity_types:
    plt.plot(grouped_data.index, grouped_data[entity_type], label=entity_type)

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Number of Businesses Registered')
plt.title('Time Series of Businesses Registered by Entity Type (Negative or Small Positive Trend)')
plt.legend()
plt.show()
```

# Correlation between Selected Metrics and Smoke Impact

This visualization explores the correlation between selected healthcare metrics and the impact of smoke over the years. The data is assumed to be stored in a DataFrame named `yearly_data` for healthcare metrics and `smoke_data` for smoke impact.

## Selected Metrics

The healthcare metrics under consideration include:

- Total Discharges
- Total Patient Days
- Total Outpatient Visits
- Total Charges
- Net Patient Revenue
- Total Operating Expense
- Total Revenue

## Data Normalization

The selected healthcare metrics are normalized to enhance comparability. Normalization involves scaling the values to a standard scale, making it easier to identify relative changes and trends.

## Smoke Impact Data

Smoke impact data is provided for each year, indicating the level of impact on the region. This data is represented in a DataFrame named `smoke_df`, with 'Year' set as the index.

## Subplots and Line Plots

Subplots are created for each selected healthcare metric, arranged in a 3x2 grid. Line plots depict the trends in normalized values over the years. Additionally, the smoke impact is represented in a separate subplot.

## Observations

The visualizations allow us to observe the relationship between healthcare metrics and smoke impact. Trends, patterns, and potential correlations can be identified, offering insights into how smoke exposure may influence healthcare indicators.

## Saving and Displaying the Plot

The generated plot is saved as an image file (`correlation_selected_metrics_smoke_impact_subplots.png`) for reference and can be included in a magazine or presentation. Additionally, the plot is displayed for immediate visualization.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'yearly_data' is a DataFrame containing the selected metrics and 'smoke_data' is the new smoke impact data
# Replace this with your actual data
selected_columns = [
    'Total Discharges',
    'Total Patient Days',
    'Total Outpatient Visits',
    'Total Charges',
    'Net Patient Revenue',
    'Total Operating Expense',
    'Total Revenue'
]

# Filter data for years 2000-2019
selected_data = yearly_data.loc[2000:2019, selected_columns]

# Normalize data for better comparison (optional)
selected_data_normalized = (selected_data - selected_data.mean()) / selected_data.std()

# Extract data from the new smoke estimator
smoke_years = smoke_data['Year']
smoke_impact = smoke_data['Smoke Impact']

# Create subplots
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(18, 12), sharex=True)
fig.suptitle('Correlation between Selected Metrics and Smoke Impact (2000-2019)', fontsize=18, y=1.02)

# Using a line plot for a smoother representation of the trend
for i, (col, ax) in enumerate(zip(selected_data_normalized.columns, axes.flatten())):
    sns.lineplot(x=selected_data_normalized.index, y=selected_data_normalized[col], label=col, ax=ax)
    ax.set_title(col, fontsize=14)
    ax.set_xlabel('Year', fontsize=12)
    ax.set_ylabel('Normalized Values', fontsize=12)
    ax.tick_params(axis='both', labelsize=10)
    ax.grid(axis='y', linestyle='--', alpha=0.7)

    # Add the new smoke impact data to the plot
    ax2 = ax.twinx()
    ax2.plot(smoke_years, smoke_impact, label='Smoke Impact', color='orange', linestyle='--')
    ax2.set_ylabel('Smoke Impact', fontsize=12)
    ax2.tick_params(axis='y', labelsize=10)
    ax2.legend(loc='upper right')

# Legend
axes.flatten()[-1].legend(loc='upper left', bbox_to_anchor=(1, 1))

plt.tight_layout()

# Save the plot as an image for the magazine
plt.savefig('../images/correlation_selected_metrics_smoke_impact_subplots.png')

# Show the plot
plt.show()
```

The visual representation delineates the connections between specific healthcare metrics and the impact of smoke during the period spanning 2000 to 2019. Summarily, the observed trends are outlined as follows:

**Total Charges:** Reflecting an upward trend, this metric implies a progressive escalation in charges associated with healthcare services, potentially influenced by factors such as inflation or enhanced service offerings.

**Net Patient Revenue:** Experiencing an increase, this metric suggests that the healthcare facility is yielding greater profits after factoring in deductions, indicating improved financial performance.

**Total Operating Expense:** In contrast to the other metrics, this particular indicator displays a decrease, signaling that the facility has succeeded in curbing its operational costs. This reduction may point towards enhanced operational efficiency and cost-effectiveness.

In the aggregate, the discerned trends imply a favorable impact of smoke on these healthcare metrics, with most metrics portraying positive growth, except for the decrease in total operating expenses.
