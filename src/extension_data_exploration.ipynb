# Number of Registered business

```python
oregon_business_data = pd.read_csv('../data/Active_Businesses_-_ALL.csv')
```

```python
oregon_business_data.dtypes
```

```python
# Get unique cities
unique_cities = oregon_business_data['City'].unique()

# Print unique cities
print("Unique Cities:")
for city in unique_cities:
    print(city)
```

```python
mcminnville_data = oregon_business_data[oregon_business_data['City'] == 'MCMINNVILLE']
```

```python
mcminnville_data
```

```python

# Assuming df is your DataFrame

# Convert 'Registry Date' to datetime format
mcminnville_data['Registry Date'] = pd.to_datetime(mcminnville_data['Registry Date'], errors='coerce')

# Filter businesses registered from 1980 to the current date
start_date = pd.to_datetime('1980-01-01')  # Adjust the start date as needed
current_date = pd.to_datetime('today')

filtered_data = mcminnville_data[(mcminnville_data['Registry Date'] >= start_date) & (mcminnville_data['Registry Date'] <= current_date)]

# Display the filtered data
print(filtered_data)
```

```python

# Convert 'Registry Date' to datetime format (if not already done)
filtered_data['Registry Date'] = pd.to_datetime(filtered_data['Registry Date'], errors='coerce')

# Extract the year from the 'Registry Date' column
filtered_data['Year'] = filtered_data['Registry Date'].dt.year

# Perform year-wise analysis (example: count of businesses registered each year)
yearly_analysis = filtered_data['Year'].value_counts().sort_index()

# Display the year-wise analysis
print("Year-wise Analysis:")
print(yearly_analysis)
```
